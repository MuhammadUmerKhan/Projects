```python
import pandas as pd
import re
```

```python
f = open('WhatsApp Chat with Hafiz Sharjeel SMIU.txt', 'r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4}, \d{1,2}:\d{2}\u202f?[APMapm]{2} - '

split_content = re.split(pattern, data)[1:]

split_content
```

```python
dates = re.findall(pattern, data)
```

```python
dates
```

```python
df = pd.DataFrame({'user_message':split_content, 'message_date':dates})
```

```python
date_format = '%m/%d/%y, %I:%M\u202f%p - '
df['message_date'] = pd.to_datetime(df['message_date'], format=date_format)
```

```python
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
user = []
messages = []
```

```python
user = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]: # user name
        user.append(entry[1])
        messages.append(entry[2])
    else:
        user.append('group_notification')
        messages.append(entry[0])
```

```python
df['user'] = user
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```
