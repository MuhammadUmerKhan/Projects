```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=433)

# Identify numeric and categorical columns
numeric_features = X.select_dtypes(include=[np.number]).columns
categorical_features = X.select_dtypes(include=[np.object_, np.bool_]).columns

# Preprocessing for numeric features
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

# Preprocessing for categorical features
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

# Define the model
model = LinearRegression()

# Create a pipeline with polynomial features
pipe = Pipeline(steps=[('preprocessor', preprocessor),
                       ('poly', PolynomialFeatures(degree=2, include_bias=False)),
                       ('model', model)])

# Fit the model
pipe.fit(X_train, y_train)

# Predict and evaluate
y_pred = pipe.predict(X_test)
print('R2 score:', r2_score(y_test, y_pred))

# Try Ridge and Lasso Regression
# Ridge
ridge = Ridge()
ridge_pipe = Pipeline(steps=[('preprocessor', preprocessor),
                             ('poly', PolynomialFeatures(degree=2, include_bias=False)),
                             ('model', ridge)])

param_grid_ridge = {'model__alpha': [0.1, 1.0, 10.0]}
ridge_search = GridSearchCV(ridge_pipe, param_grid_ridge, cv=5, scoring='r2')
ridge_search.fit(X_train, y_train)
y_pred_ridge = ridge_search.predict(X_test)
print('R2 score (Ridge):', r2_score(y_test, y_pred_ridge))

# Lasso
lasso = Lasso()
lasso_pipe = Pipeline(steps=[('preprocessor', preprocessor),
                             ('poly', PolynomialFeatures(degree=2, include_bias=False)),
                             ('model', lasso)])

param_grid_lasso = {'model__alpha': [0.01, 0.1, 1.0]}
lasso_search = GridSearchCV(lasso_pipe, param_grid_lasso, cv=5, scoring='r2')
lasso_search.fit(X_train, y_train)
y_pred_lasso = lasso_search.predict(X_test)
print('R2 score (Lasso):', r2_score(y_test, y_pred_lasso))
```
